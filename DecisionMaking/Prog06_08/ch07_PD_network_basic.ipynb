## NetworkX（グラフ用ライブラリ）のインポート

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## 定数の設定

```python
NUM_AGENT = 50 #エージェント（ノード）数
```

## ランダムネットワーク

```python
G = nx.gnp_random_graph(NUM_AGENT, 0.3)
pos = nx.circular_layout(G)
plt.subplots(figsize = (6, 6))
nx.draw_networkx(G, pos,node_color="lightgray", edgecolors='k', edge_color='lightgrey', node_size=50, with_labels=False)
plt.savefig("gnp_random_graph.pdf")
```

## スケールフリーネットワーク

```python
G = nx.barabasi_albert_graph(n = NUM_AGENT, m=1)
pos = nx.kamada_kawai_layout(G)
plt.subplots(figsize = (6, 6))
nx.draw_networkx(G, pos,node_color="lightgray", edgecolors='k', edge_color='lightgrey', node_size=50, with_labels=False)
plt.savefig("barabasi_albert_graph.pdf")
```
