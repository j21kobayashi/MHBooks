### 選好の定義域(domain of preference)

選択肢の集合が $ O = \{a, b, c\} $ のときにあり得る全ての選好順序を列挙する

```python
import itertools
import math
from IPython.display import display_markdown

alternatives = ('a', 'b', 'c') 
perms = tuple(itertools.permutations(alternatives))
pref_str = lambda p: r' $\succ$ '.join(p)
display_markdown(',&emsp;'.join(pref_str(p) for p in perms), raw=True)
print('Number of preferences: ', len(perms))
print(f'{len(alternatives)}! = {math.factorial(len(alternatives))}')
```

選択肢の集合が $ O = \{a, b, c\} $ かつ投票者の人数が $|N| = n = 3$ のときにあり得る全ての選好プロファイルを
列挙する。

これらの選好プロファイル全体の集合を $L^n$ とするとき，「任意のプロファイル」は $\forall [\succ] \in L^n$ のように表現できる。

```python
import numpy as np
import itertools as it
from IPython.display import display_markdown

alternatives = ('a', 'b', 'c') 
perms = tuple(it.permutations(alternatives))
n = 3
prof = np.array(tuple(it.product(*it.repeat(perms, n))))
pref_str = lambda p: r' $\succ$ '.join(p)
print(prof.shape)
display_markdown('  \n'.join(f'{i}:&emsp;' + ',&emsp;'.join(pref_str(p) for p in prefs) for i,prefs in enumerate(prof)), raw=True)
```
