#### 情報エントロピー
- ベルヌーイ分布 

    $$
    \begin{aligned}
        s_i(a_i) = q_i^{x(a_i)} (1 - q_i)^{1 - x(a_i)} 
    \end{aligned}
    $$

    について，情報エントロピーは $H[a_i] = E[-\log_b s_i(a_i)]$  と表される。

- ここで，$x(a_i)$ は次の関数である。

    $$
    \begin{aligned}
        x(a_i) = 
        \begin{cases}
            1 \quad \text{if} \quad a_i = \text{L}  \\
            0 \quad \text{if} \quad a_i = \text{R} 
        \end{cases}
    \end{aligned}
    $$

   ただし，$ \text{L}, \text{R}$ はラベル (定数) であり，変数ではない。

-    $q_i \in [0, 1]$は確率分布を調整するパラメータであり，$q_i = 1$
    に近いほど$a_i = \text{L}$となる事象が発生しやすくなり，
    $q_i = 0$に近いほど$a_i = \text{R}$となる事象が発生しやすくなる。

- 以下では，$i = 1$ に注目し，$q_1 = q$ と置いて，$H[a_1]$ と $q$ の関係をプロットする。

    $$
    \begin{aligned} 
    H[a_1] & =  E[-\log_b s_1(a_1)] \\
    {}     & =  -s_1(a_1=\text{L})\log_b s_1(a_1=\text{L}) -s_1(a_1=\text{R})\log_b s_1(a_1=\text{R}) \\
    {}     & =  -q \log_b q - (1-q) \log_b (1-q)
    \end{aligned}
    $$
    
- $H[a_1]$ の単位を bit にするため，対数の底を $b = 2$ とする。

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 

# mpl.style.use('grayscale') 
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['ytick.labelsize'] = 'x-large' #'large'
plt.rcParams['xtick.labelsize'] = 'x-large' #'large'
plt.rcParams['axes.labelsize'] = 'x-large' #'large'
plt.rcParams['axes.titlesize'] = 'x-large' #'large'

def h(q): 
    return -q*np.log2(q) - (1-q)*np.log2(1-q)

q = np.linspace(0.01, 0.99, 100)

plt.plot(q, h(q))
plt.xlabel("$q$", fontsize='xx-large')
plt.ylabel("$H[a_1]$", fontsize='xx-large')
plt.grid(linestyle=':')
# plt.savefig("entropy_baba-nuki.pdf", bbox_inches='tight')
```
