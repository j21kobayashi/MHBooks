## 収入同値定理に関連する各種の導出

### 第1価格秘密入札: ナッシュ均衡における支払額の期待値 $E[\wp]$
参加者の真の評価額 $\{v_i\}_{i=0}^n$ が，全て独立に $[0, 1]$ 上の一様分布に従うとする。
$v_i$ を小さい順に並べなおしたものを $v_{(k)}$ と表す。
$$
    v_{(1)} \leq v_{(2)} \leq \ldots \leq v_{(k)} \leq \ldots \leq v_{(n-1)} \leq v_{(n)}
$$

$v_{(k)}$ は順序統計量と呼ばれ，書籍第11章において示したように，
$k$ 番目の順序統計量 $v_{(k)}$ の確率密度関数は次のように表される。
$$
\begin{aligned}
   f_{v_{(k)}}(v) & = \dfrac{n!}{(k-1)!(n-k)!} v^{k-1}(1-v)^{n-k} \\
                  & = \dfrac{1}{B(k, n-k+1)} v^{k-1}(1-v)^{n-k} 
\end{aligned}
$$
ここで，$B(a, b)$ はベータ関数である。

したがって，上記の確率密度関数 $f_{v_{(k)}}(v)$ に従う確率変数 $v$ の関数 $g(v)$ の
期待値は次式で計算できる。
$$ 
\begin{aligned}
E_{v_{(k)}}[g(v)] & = \int_0^1 g(v) f_{v_{(k)}}(v) dv \\
        & = \int_0^1 g(v) \dfrac{n!}{(k-1)!(n-k)!} v^{k-1}(1-v)^{n-k} dv
\end{aligned}
$$

一方，真の評価額が $v$ のとき，ナッシュ均衡戦略の申告額は $\hat{v} = \dfrac{n-1}{n} v$ 
である。

したがって，第1価格，すなわち 最大の申告額 $\hat{v}_{(n)} = \dfrac{n-1}{n} v_{(n)}$ の期待値は
次のように求められる。

<a name="eq-exp-1st-price"> </a> 
$$ 
\begin{aligned}
E_{v_{(n)}}[\hat{v}] & = \int_0^1 \dfrac{n-1}{n} v f_{v_{(n)}}(v) dv \\
        & = \int_0^1 \dfrac{n-1}{n} v \dfrac{n!}{(n-1)!(n-n)!} v^{n-1}(1-v)^{n-n} dv \\
        & = \int_0^1 \dfrac{n-1}{n} v n v^{n-1}(1-v)^{0} dv \\
        & = (n-1) \int_0^1  v^{n} dv \\
        & = (n-1) \left[ \dfrac{v^{n+1}}{n+1} \right]_0^1 \\
        & = (n-1) \left( \dfrac{1}{n+1} - 0 \right) \\
        & = \dfrac{n-1}{n+1}
\end{aligned}
$$

第1価格秘密入札において，その定義から，落札者の支払額 $\wp$ は第1価格なので，
支払額 $\wp$ の期待値 $E[\wp]$ は，第1価格の期待値 $E_{v_{(n)}}[\hat{v}]$ に一致する。

$$
E[\wp] = E_{v_{(n)}}[\hat{v}] = \dfrac{n-1}{n+1}
$$

### 第1価格秘密入札: ナッシュ均衡における支払額の分散 $V[\wp]$ 

分散の定義から，次式が成立する。

$$
V[\wp] = E[\left(\wp - E[\wp]\right)^2] = E[\wp^2] - \left(E[\wp]\right)^2
$$

前節に示したように，落札者の支払額は第1価格，すなわち最大の申告額 $v_{(n)}$ なので，ナッシュ均衡において次式が成立する。
$$
V[\wp] = E[\wp^2] - \left(E[\wp]\right)^2 = E_{v_{(n)}}[\hat{v}^2] - \left( E_{v_{(n)}}[\hat{v}] \right)^2
$$

このうち，最右辺の第2項の $E_{v_{(n)}}[\hat{v}]$ は[前節](#eq-exp-1st-price)で導出済みなので，第1項の $ E_{v_{(n)}}[\hat{v}^2] $ を求める。

$$
\begin{aligned}
E_{v_{(n)}}[\hat{v}^2] 
        & = \int_0^1 \left( \dfrac{n-1}{n} v \right)^2 f_{v_{(n)}}(v) dv \\
        & = \int_0^1 \dfrac{(n-1)^2}{n^2} v^2 \dfrac{n!}{(n-1)!(n-n)!} v^{n-1}(1-v)^{n-n} dv \\
        & = \int_0^1 \dfrac{(n-1)^2}{n^2} v^2 n v^{n-1}(1-v)^{0} dv \\
        & = \dfrac{(n-1)^2}{n^2} n \int_0^1  v^2  v^{n-1} dv \\
        & = \dfrac{(n-1)^2}{n} \int_0^1 v^{n+1} dv \\
        & = \dfrac{(n-1)^2}{n} \left[ \dfrac{v^{n+2}}{n+2} \right]_0^1 \\
        & = \dfrac{(n-1)^2}{n}  \left( \dfrac{1}{n+2} - 0 \right) \\
        & = \dfrac{(n-1)^2}{n(n+2)} \\
\end{aligned}
$$

この結果を用いれば，第1価格秘密入札のナッシュ均衡における支払額の分散は次のように
なる。

$$
\begin{aligned}
V[\wp] &= E_{v_{(n)}}[\hat{v}^2] - \left( E_{v_{(n)}}[\hat{v}] \right)^2 \\
       & = \dfrac{(n-1)^2}{n(n+2)} - \dfrac{(n-1)^2}{(n+1)^2} \\
       & = \dfrac{(n-1)^2 (n+1)^2 - (n-1)^2 n(n+2) }{n(n+1)^2(n+2)} \\
       & = \dfrac{(n-1)^2 \left( (n+1)^2 - n(n+2) \right)}{n(n+1)^2(n+2)} \\
       & = \dfrac{(n-1)^2 \left( (n^2 + 2n + 1) - (n^2 + 2n) \right)}{n(n+1)^2(n+2)} \\
       & = \dfrac{(n-1)^2 }{n(n+1)^2(n+2)} \triangleq V_{\mathrm{1st}}[\wp] \\
\end{aligned}
$$

### 第2価格秘密入札: 支配戦略均衡における支払額の期待値 $E[\wp]$

真の評価額が $v$ のとき，第2価格秘密入札における支配戦略の申告額は 
$\hat{v} = v$ である（正直な申告が最良の戦略となる）。

一方，参加者の人数が $n$ 人の第2価格秘密入札において，その定義から，支払額 $\wp$ となる第2価格は
2番目に大きい申告額，すなわち， $(n-1)$ 番目に小さい申告額 $\hat{v}_{(n-1)}$ である。
$$
    v_{(1)} \leq v_{(2)} \leq \ldots \leq v_{(k)} \leq \ldots \leq v_{(n-1)} \leq v_{(n)}
$$

支配戦略均衡においては，全員の申告額が真の評価額に一致し，
すなわち $\forall i \in N$: $\hat{v}_i = v_i$ なので，第2価格 $\hat{v}_{(n-1)}$ は，$(n-1)$ 番目の順序統計量
$v_{(n-1)}$ そのものとなる。
$$
    \hat{v}_{(n-1)} = v_{(n-1)}
$$

したがって，支配戦略均衡における支払額の期待値は次の関係を持つ。
$$
 E[\wp] = E_{v_{(n-1)}}[\hat{v}] = E_{v_{(n-1)}}[v]
$$

書籍第11章における議論から，$k$番目の順序統計量 $v_{(k)}$ の期待値は次のように求められる。
$$ 
\begin{aligned}
E_{v_{(k)}}[v] & = \int_0^1 v f_{v_{(k)}}(v) dv \\
        & = \int_0^1 v \dfrac{n!}{(k-1)!(n-k)!} v^{k-1}(1-v)^{n-k} dv \\
  &  = \dfrac{n!}{(k-1)!(n-k)!} B(k+1, n-k+1) \\
  &  = \dfrac{n!}{(k-1)!(n-k)!} \dfrac{\Gamma(k+1)\Gamma(n-k+1)}{\Gamma((k+1) + (n-k+1))} \\
  &  = \dfrac{n!}{(k-1)!(n-k)!} \dfrac{k!(n-k)!}{(n+1)!} \\
  &  = \dfrac{k}{n+1} 
\end{aligned}
$$
ただし，途中でベータ関数とガンマ関数の
関係式
$$
\Gamma(a)\Gamma(b) = \Gamma(a + b)B(a, b)
$$
および 
$$
\Gamma(n + 1) = n!
$$
を利用した。

したがって，支配戦略均衡における支払額の期待値 $E[\wp]$ は以下のとおりとなる。
$$
E[\wp] = E_{v_{(n-1)}}[v] = \dfrac{n-1}{n+1}
$$

これは，第1価格秘密入札のナッシュ均衡における[支払額の期待値](#eq-exp-1st-price)と一致している。

### 第2価格秘密入札: 支配戦略均衡における支払額の分散 $V[\wp]$

第1価格秘密入札の場合と同様に，支払額の分散は次のように表される。

$$
V[\wp] = E[\wp^2] - \left(E[\wp]\right)^2 = E_{v_{(n-1)}}[v^2] - \left( E_{v_{(n-1)}}[v] \right)^2
$$


このうち，最右辺の第2項の $ E_{v_{(n-1)}}[v] $ は前節で導出済みなので，第1項の $ E_{v_{(n-1)}}[v^2] $ を求める。

$$
\begin{aligned}
E_{v_{(n-1)}}[v^2]  
    & = \int_0^1 v^2 f_{v_{(n-1)}}(v) dv \\
    & = \int_0^1 v^2 \dfrac{n!}{((n-1)-1)!(n-(n-1))!} v^{(n-1)-1}(1-v)^{n-(n-1)} dv \\
    & = \int_0^1 v^2 \dfrac{n!}{(n-2)!1!} v^{n-2}(1-v)^{1} dv \\
    & = n(n-1) \int_0^1 v^n(1-v) dv \\
    & = n(n-1) \int_0^1 v^n -v^{n+1} dv\\
    & = n(n-1) \left[ \dfrac{v^{n+1}}{n+1} -\dfrac{v^{n+2}}{n+2} \right]_0^1 \\
    & = n(n-1) \left( \dfrac{1}{n+1} -\dfrac{1}{n+2} - 0\right) \\
    & = n(n-1)  \dfrac{(n+2) - (n+1)}{(n+1)(n+2)} \\
    & =   \dfrac{n(n-1)}{(n+1)(n+2)} \\
\end{aligned}
$$


この結果を用いれば，支配戦略均衡における支払額の分散は次のように求められる。

$$
\begin{aligned}
V[\wp] 
    & = E_{v_{(n-1)}}[v^2] - \left( E_{v_{(n-1)}}[v] \right)^2 \\
    & =  \dfrac{n(n-1)}{(n+1)(n+2)}  - \dfrac{(n-1)^2}{(n+1)^2} \\
    & =  \dfrac{n-1}{n+1} \left( \dfrac{n}{n+2}  - \dfrac{n-1}{n+1} \right)\\
    & =  \dfrac{n-1}{n+1} \cdot \dfrac{n(n+1) -(n-1)(n+2) }{(n+1)(n+2)} \\
    & =  \dfrac{n-1}{n+1} \cdot \dfrac{n^2 + n -(n^2 + n - 2) }{(n+1)(n+2)} \\
    & =  \dfrac{n-1}{n+1} \cdot \dfrac{ 2}{(n+1)(n+2)} \\
    & =  \dfrac{ 2 (n - 1)}{(n+1)^2(n+2)} \triangleq V_{\mathrm{2nd}}[\wp] \\
\end{aligned}
$$

### 収入同値の定理の条件(2) に関する中間期待効用の式の導出

以下に書籍11章において示した次式の導出過程を示す。
$$
    u_i(v_i) = u_i(\underline{v}) + \int_{x = \underline{v}}^{v_i} P_i(x) dx 
$$

書籍11章の議論から，参加者全員が均衡戦略をとるときには，どのオークションプロトコルについても
パレート効率性が実現されている。

このとき，真の評価額が$v_i$である参加者$i$が財の配分を受ける確率は次のように書き直せる。
$$
P_i(v_i) \triangleq P(v_i = \max_{j \in N} v_j \mid v_i)  
$$

$P_i(v_i)$ を用いると，真の評価額が $v_i$ のときの参加者 $i$ の中間期待効用 $EU_i(s, v_i)$

$$
\begin{aligned}
    EU_i(s, v_i) 
    & = E_{v_{-i} \mid v_i}[u_i(o, (v_i, v_{-i}))] \\
    & = E_{v_{-i} \mid v_i}[\tilde{u}_i(x, (v_i, v_{-i})) - \wp_i]  \\
    & = E_{v_{-i} \mid v_i}[\tilde{u}_i(x, (v_i, v_{-i}))] - E_{v_{-i} \mid v_i}[\wp_i] 
\end{aligned}
$$

$v_i$ が与えられたとしても，残りの $v_{-i}$ については不確定性がある。
$v_{-i}$ は様々な値をとりうるが，中間期待効用の観点では次の二つに場合分けできる。

- 参加者 $i$ が財の配分を受ける場合: このとき，$i$ の効用は $v_i$
- 参加者 $i$ が財の配分を受けない場合: このとき，$i$ の効用は 0

したがって，中間期待効用はさらに，次のように変形される。
$$
\begin{aligned}
    EU_i(s, v_i) 
    & = v_i P_i(v_i) + 0 \cdot (1 - P_i(v_i)) - E_{v_{-i} \mid v_i}[\wp_i] \\
    & = v_i P_i(v_i) - E_{v_{-i} \mid v_i}[\wp_i] \triangleq u_i(v_i) 
\end{aligned}
$$

次に真の評価額 $v_i$ の微小な変動に対応する効用の変化を調べるため，
真の評価額が $v_i$ のときに，敢えて
真の評価額が $v_i$ とわずかに異なる $ \tilde{v}_i = v_i + \Delta v_i $ の場合の均衡戦略をとることを考える。
この場合の中間期待効用 $u_i^{\prime}(v_i)$ は次のように書ける。

$$
    u_i^{\prime}(v_i) = v_i P_i(\tilde{v}_i ) - E_{v_{-i} \mid \tilde{v}i}[\tilde{\wp}_i]
$$

つまり，勝敗に関する確率は $P_i(\tilde{v}_i )$ に置き換わり，支払額に関する期待値も
真の評価額が $ \tilde{v}_i $ の場合の均衡戦略に対応するものとなる。

これに対して，真の評価額が $ \tilde{v}_i $ そのもののときに，
素直に均衡戦略をとると，この場合の中間期待効用 $u_i(\tilde{v}_i)$ は次のように書ける。

$$
    u_i(\tilde{v}_i) = \tilde{v}_i P_i(\tilde{v}_i ) - E_{v_{-i} \mid \tilde{v}_i}[\tilde{\wp}_i]
$$

これらの二つの場合の期待効用の差をとれば，支払額の期待値の項（未知）を消すことができる。
$$
\begin{aligned}
    u_i^{\prime}(v_i) - u_i(\tilde{v}_i) = (v_i - \tilde{v}_i) P_i(\tilde{v}_i ) \\
    \rightarrow 
    u_i^{\prime}(v_i) = u_i(\tilde{v}_i) + (v_i - \tilde{v}_i) P_i(\tilde{v}_i )
\end{aligned}
$$

一方，均衡戦略の定義から，真の均衡戦略から外れたときの効用 $u_i^{\prime}(v_i)$
真の均衡戦略をとるときの効用 $u_i(v_i)$ を上回ることがない。
したがって，次の不等式が成立する。
$$
\begin{aligned}
u_i(v_i) & \geq u_i^{\prime}(v_i) \\
         & \geq u_i(\tilde{v}_i) + (v_i - \tilde{v}_i) P_i(\tilde{v}_i )
\end{aligned}
$$

さらに，上記の不等式に $ \tilde{v}_i = v_i + \Delta v_i$ と，これを変形した 
$v_i - \tilde{v}_i = - \Delta v_i$ を代入すると，次のようになる。
$$
\begin{aligned}
u_i(v_i) & \geq u_i(v_i + \Delta v_i) - \Delta v_i P_i(v_i + \Delta v_i) \\
\rightarrow \Delta v_i P_i(v_i + \Delta v_i) & \geq u_i(v_i + \Delta v_i) - u_i(v_i)
\end{aligned}
$$

以上と同様の議論を $\tilde{v}_i$ と $v_i$ を入れ替えて検討すれば，次の不等式が成立することがわかる。
$$
\begin{aligned}
u_i(v_i + \Delta v_i) & \geq u_i(v_i) + \Delta v_i P_i(v_i) \\
\rightarrow u_i(v_i + \Delta v_i) - u_i(v_i) & \geq \Delta v_i P_i(v_i)
\end{aligned}
$$

以上の二つの不等式をまとめると，次のようになる。
$$
 P_i(v_i + \Delta v_i) \geq \dfrac{u_i(v_i + \Delta v_i) - u_i(v_i)}{\Delta v_i} \geq P_i(v_i)
$$

この不等式について，$\Delta v_i \rightarrow 0$ の極限をとれば，はさみうちの原理（squeeze theorem）により次式を得る。
$$
\dfrac{d u_i(v_i)}{d v_i} = P_i(v_i)
$$

この式を積分すると，冒頭の式が得られる。
$$
    u_i(v_i) = u_i(\underline{v}) + \int_{x = \underline{v}}^{v_i} P_i(x) dx 
$$

参加者全員が均衡戦略をとるとき，第1・2価格秘密入札のいずれについても，真の評価額が 
最小の $\underline{v}$ である参加者が財の配分を受けることはない。
したがって，$u_i(\underline{v}) = 0$ が成立する。
このとき，
$$
    u_i(v_i) = \int_{x = \underline{v}}^{v_i} P_i(x) dx 
$$
となり，$u_i(v_i)$ は第1・2価格秘密入札のいずれについても，同一となる。

改めて $u_i(v_i)$ の定義を次のように変形する。
$$
\begin{aligned}
    u_i(v_i) = v_i P_i(v_i) - E_{v_{-i} \mid v_i}[\wp_i]  \\
    \rightarrow  E_{v_{-i} \mid v_i}[\wp_i] = v_i P_i(v_i) - u_i(v_i)
\end{aligned}
$$
これまでの議論から，右辺の第1項，第2項は，第1・2価格秘密入札のいずれにおいても
同一であり，したがって，支払額の期待値も両方のプロトコルで同一となる。
支払額はそのまま主催者の収入となるため，収入の期待値も第1・2価格秘密入札の
いずれにおいても同一となる。
